# Análisis De Datos

```python
# Pasajero 148
df[df['PassengerId']=='148']
```

```python
# Total pagado
df["Fare"].sum()
```

```python
# Sí Sobrevivió
df[df['Survived'] == '1'].head()
```

```python
# No Sobrevivió
df[df['Survived'] == '0'].head()
```

```python
# Porcentaje (Respecto Total Pasajeros) de muertos y superviviente
round((df["Survived"].replace(["0","1"],["% de Muertos","% de Vivos"]).value_counts() * 100 )/len(df),2).astype(str) + " %"
```

```python
# Nombres en Orden Alfabético de la Primera Clase
df[df['Pclass'] == '1']["Name"].sort_values()
```

```python
# Edad promedio de las mujeres que viajaban en cada clase
round(df[df['Sex'] == 'female'].groupby("Pclass")["Age"].mean())
```

## Modificación De Datos

```python
df["MenorEdad"] = (df['Age'] < 18)

# Porcentaje (Respecto Total Pasajeros) de menores y mayores de edad que sobrevivieron en cada clase
round((df.groupby(["Pclass", "MenorEdad"])["Survived"].value_counts() * 100 )/len(df),2).astype(str) + " %"
```
